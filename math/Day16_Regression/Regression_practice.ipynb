```python
from sklearn.datasets import make_regression
```

```python
X, y, w = make_regression(
    n_samples=50, n_features=1, bias=100, noise = 10, coef = True, random_state=0
)
```

```python
xx = np.linspace(-3, 3, 100)
y0 = w * xx + 100
plt.plot(xx, y0, "r-")
plt.scatter(X, y, s=100)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
%matplotlib inline
```

```python
def make_regressoin2(n_samples, bias, epsilon, random_state=0):
    np.random.seed(random_state)
    X = np.random.normal(size = n_samples) # sp.stats.norm()
    w = 100 * np.random.normal(size=10)[0]
    epsilon = np.random.normal(scale = epsilon, size=n_samples)
    y0 = X * w + bias
    y = y0 + epsilon
    return X, y, w
```

```python
X, y, w = make_regressoin2(50, 100, 10 )
```

```python
plt.scatter(X, y, s = 100)
plt.show()
%matplotlib inline
```
