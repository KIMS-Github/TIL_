### INDICATOR

```python
n = np.linspace(-5, 5, 10)
```

```python
def stepfunc(x):
    if isinstance(x, np.ndarray):
        return np.where(x >= 0, 1, 0)
    else:
        
        return 1.0 if x >= 0 else 0.0
    
```

```python
stepfunc = stepfunc(n)
```

```python
plt.plot(n, stepfunc)
plt.ylim(-0.5, 1.5)
plt.show()
%matplotlib inline
```

# 로그함수

### 로그함수는 지수함수의 역함수

# 로그함수가 중요한 이유
## 1. 곱하기를 더하기로 바꾼다 = 어떤 곱에 로그는 각각의 로그의 합과 같다.


$$ 
\begin{align}
\log{(x_1 \cdot x_2)} = \log{x_1} + \log{x_2} 
\tag{4.1.30}
\end{align}
$$

이 성질을 확장하면 다음과 같이 표기한다.

$$ 
\begin{align}
\log{\left(\prod_i x_i\right)} = \sum_i \left(\log{x_i}\right) 
\tag{4.1.31}
\end{align}
$$


위 식에서 $x_1 = x_2 = \cdots = x$ 라면

$$ 
\begin{align}
\log x^n = n \log x 
\tag{4.1.32}
\end{align}
$$

```python
def softmax(x,w):
    e = np.exp(x*w)
    return e/e.sum()
```
