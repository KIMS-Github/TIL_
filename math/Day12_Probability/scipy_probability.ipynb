# 베르누이 분포와 이항분포
Bernoulli trial and Binomial

베르누이 시행(두가지중 하나 예:동전)의 결과값을 가지는 베르누이 확률변수 : 숫자가된다

$$
\begin{align}
\text{Bern}(x;\mu) = \mu^x(1-\mu)^{(1-x)}
\tag{8.2.3}
\end{align}
$$

```python
µ=0.6
rv = sp.stats.bernoulli(µ)
type(rv)
```

```python
xx = [0 , 1]
plt.bar(xx, rv.pmf(xx))
plt.xticks([0, 1], ["x=0", "x=1"])
plt.show()
%matplotlib inline
```

```python
samples = rv.rvs(100, random_state=0) # 동전을 100번 던졌을때 결과와 같음
samples
```

```python
sns.countplot(samples)
```

# 이항분포
동전을 N번 던져서 앞면이 한번도 안나오는 0번 부터 10번다 다 나오는 
{0, 1, 2, ..., 10} 총 11개의 표본값을 갖는 분포
## Binomial
1번던지면 베르누이 N번 던지면 이항분포

$$
\begin{align}
X \sim \text{Bin}(x;N,\mu)
\tag{8.2.11}
\end{align}
$$

예를 들어 동전을 N번 던져 앞면이 나오는 포본 0~ 10 이 x 가 된다.

$$
\begin{align}
x = \sum_{i=1}^N y_i
\tag{8.2.13}
\end{align}
$$


위의 식에서 확률변수를 다 더하면 카운트 한것과 같다. 

# Scipy 로 구현하는 Binomial

```python
N = 10
µ = 0.6
rv = sp.stats.binom(N, µ)
```

```python
xx = np.arange(N+1)
plt.bar(xx, rv.pmf(xx), align = "center")
plt.show()
# µ * N 값이 최빈값 (mode) 가 된다
```
