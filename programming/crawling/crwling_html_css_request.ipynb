```python
import requests
from bs4 import BeautifulSoup

# network tab "ALL"

query = "삼성전자"
url = "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query={}".format(query)
print(url)
```

```python
# requests and response : HTML(str)
response = requests.get(url)
response
```

```python
response.text
```

```python
doc = BeautifulSoup(response.content, "html.parser")
type(doc)
```

```python
# doc
# select: 여러개 엘리먼트 가져옴
# select_one: 한개의 엘리먼트 가져옴
```

```python
# li 태그를 사용해서 가져올려고 했더니 너무 많은 태그가 있음
elements = doc.select('._related_keyword_ul > li')
len(elements)
```

```python
# chrome copy selector use 단, 계층구조의 상위 id 까지 찾아줌
#nx_related_keywords > dl > dd.lst_relate._related_keyword_list > ul 
```

```python
keywords=[]
for element in elements:
    keywords.append(element.text.strip())
    
print(keywords)
```

```python
# 위의 코드 list comprehensionize 
keywords = [_.text.strip() for _ in elements]
print(keywords)
```

```python
def related_search_keywords(query):
    
    url = "https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query={}".format(query)
    
    response = requests.get(url)
    
    doc = BeautifulSoup(response.content, "html.parser")
    
    elements = doc.select('._related_keyword_ul > li')
    
    keywords = [_.text.strip() for _ in elements]
    
    return keywords
    
```

```python
related_search_keywords('covid')
```
