```python
raw_df = pd.read_excel("resources/일일등급속보_20200609.xls")

raw_df.head()
```

```python
def screen_name(df):
    df = df[[df.columns[0],'Unnamed: 2','Unnamed: 4','Unnamed: 6']]
    df.columns = df.loc[0]
    df = df.drop([0,1])
    df["등급결정일"] = df["등급결정일"].apply(lambda x: int(str(x).replace(".","")))
    
    return df
```
